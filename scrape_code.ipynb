This program is to implement web scraping on toscrape.com site and create DataFrame with author name and Quote as columns from scraped data

Conditions:-
    ->extract only Quotes and authors name which belongs to tag type "life"
    ->run automated script to send request to site only once , download data to file , read from file , read all pages till the end.
     


```python
from bs4 import BeautifulSoup
import requests

```

```python
import pandas as pd
df=pd.DataFrame(columns=["author","quotes"])
df
```

```python
pg_no=1
while True:
    
    URL=f"https://quotes.toscrape.com/page/{pg_no}/"
    page=requests.get(URL).text
    soup = BeautifulSoup(page, "html.parser")
    check=soup.select("div.quote") 
    """The website does not return error code for entering web page no. that does not exist
        to solve this we look for any existing quotes inside a given page to conform if page exists"""
    if not check:
        print("pages over!")
        break
    with open(f"pages/pageno{pg_no}.html", "w", encoding="utf-8") as f:
        f.write(page)
        print(f"downloaded data form page=>{pg_no}")
        pg_no+=1
```

```python

pg_no=1
count=0
while True:
    try:
        with open(f"pages/pageno{pg_no}.html", "r", encoding="utf-8") as f:
            html_pages=f.read()
            pg_no+=1
        
        soup=BeautifulSoup(html_pages,"html.parser")
        
        all_quotes=soup.select("div.quote") 
        
        for q in all_quotes:
            only_tags=q.select(".tags .tag")
            for tag in only_tags:

                if tag.get_text() =="life":
                    author=q.select("span small.author")[0].get_text()
                    quote=q.select("span.text")[0].text
                    df.loc[count]=[author,quote]
                    count+=1
            
        
    except FileNotFoundError:
        print("no more files")
        break







```
